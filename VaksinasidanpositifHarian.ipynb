```python
data_plot5 = pd.DataFrame({"Tanggal" : Vak['Tanggal'],
                           "Positif H+14" : Vak['Positif H+14'], 
                           "Vaksin 1": Vak['Total Vak 1'], "Vaksin 2": Vak['Total Vak 2'], 
                           'Vak 1 Harian' : Vak['diff Vak 1'], 
                           'Vak 2 Harian' : Vak['diff Vak 2'],
                           'Total Vak Harian' : Vak['Total diff Vak'],
                           'Positif H+14' : Vak['Positif H+14'],
                           'Positif H+7' : Vak['Positif H+7']})
data_plot5
```

```python
# data_plot5['Tanggal'] = pd.to_datetime(data_plot5['Tanggal'])
# data_plot5['Tanggal2'] = data_plot5['Tanggal'].data_plot5.strftime('%m/%d/%Y')
data_plot5['Tanggal'] = pd.to_datetime(data_plot5['Tanggal'], format='%d-%m-%y')
data_plot5
```

```python
dateList = []
b = 0
for i in Vak.Tanggal:
    a = str(i)
    if(a[8:10] == '01'):
        dateList.append(i)
```

```python
plt.figure(figsize=(10,7))
width = 2
# plt.plot(data_plot5.Tanggal,data_plot5['Vak 1 Harian'],label = "Vak 1")
# plt.plot(data_plot5.Tanggal,data_plot5['Vak 2 Harian'],label = "Vak 2")
plt.plot(data_plot5.Tanggal,data_plot5['Vak 1 Harian'].rolling(width, win_type='triang').sum(),label = "Vaksinasi 1 Harian")
plt.plot(data_plot5.Tanggal,data_plot5['Vak 2 Harian'].rolling(width, win_type='triang').sum(),label = "Vaksinasi 2 Harian")
# plt.plot(data_plot5.Tanggal,data_plot5['Total Vak Harian'].rolling(width, win_type='triang').sum() ,label = "Total Vaksin Harian")
plt.plot(data_plot5.Tanggal,data_plot5['Total Vak Harian'] ,label = "Total Vaksin Harian")
# plt.plot(data_plot5.Tanggal,data_plot5['Positif H+14'], label = "Positif H+14")
plt.plot(data_plot5.Tanggal,data_plot5['Positif H+7'], label = "Positif H+7")
plt.xticks(dateList)
plt.grid()
plt.legend()
plt.show()
```
